```python
import datasets
import pandas as pd
import os

import warnings
warnings.filterwarnings(action='once')
```

```python
df_goemotions = datasets.load_dataset("go_emotions")
df_goemotions
```

```python
df_train = pd.DataFrame(df_goemotions['train'])
df_test = pd.DataFrame(df_goemotions['test'])
df_val = pd.DataFrame(df_goemotions['validation'])
new_df = pd.concat([df_train, df_test, df_val])
new_df
```

```python
new_df['labels'] = new_df['labels'].apply(lambda x: x[0])
new_df.drop(['id'], axis=1, inplace=True)
new = new_df[new_df['labels'] == 27][:5000]
new['labels'] = 0
new
```

```python
sms_df = pd.read_csv('./training_data/SMSSpamCollection', sep='\t', header=None)
sms_df.columns=['labels', 'text']
sms_df
```

```python
sms_df['labels'] = sms_df['labels'].replace({'ham': 0, 'spam': 1})
sms_df.drop_duplicates(inplace=True)
sms_df
```

```python
youtube_df = pd.DataFrame()
for file_name in os.listdir('training_data/youtube_comments'):
    youtube_df = pd.concat([youtube_df, pd.read_csv(f'training_data/youtube_comments/{file_name}')])
youtube_df
```

```python
youtube_df.drop(['COMMENT_ID', 'AUTHOR', 'DATE'],axis=1, inplace=True)
youtube_df.columns = ['text', 'labels']
youtube_df.drop_duplicates(inplace=True)
youtube_df['labels'] = pd.to_numeric(youtube_df['labels'])
youtube_df
```

```python
spam_df = pd.concat([youtube_df, sms_df, new])
spam_df = spam_df[spam_df['text'].duplicated(keep=False) == False]
spam_df
```

```python
spam_df.to_csv('./training_data/my_spam_dataset.csv', index=False)
```
